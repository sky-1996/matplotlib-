# 完善原始折线图（辅助显示层）

```python
#画出一小时内每分钟的温度变化折线图，变化范围在15-18度之间
import random
#准备数据x,y
x=range(60)
y1=[random.uniform(15,18)for i in x]
y2=[random.uniform(1,3)for i in x]
#创建画布
plt.figure(figsize=(20,8),dpi=80)
#绘制图像
plt.plot(x,y1,color='r',linestyle='-.',label='上海')
plt.plot(x,y2,color='g',linestyle='--',label='北京')
#显示图例
plt.legend()
#修改x,y刻度
x_label = ["11点{}分".format(i) for i in x]
plt.xticks(x[::5],x_label[::5])
plt.yticks(range(0,30,3))
#添加网格显示
plt.grid(linestyle='--',alpha=0.5)
#完善坐标轴信息
plt.xlabel('时间')
plt.ylabel('温度(单位：℃)')
plt.title('两个城市一小时内温度变化')
#显示图像
plt.show()
```

多个坐标显示-plt.subplots(面向对象的画图方法)

```python
#准备数据
x=range(60)
y1=[random.uniform(15,18)for i in x]
y2=[random.uniform(1,3)for i in x]
#创建绘图区
figure,axes=plt.subplots(nrows=1,ncols=2,figsize=(20,8),dpi=80)
#绘制图像
axes[0].plot(x,y1,color='r',linestyle='--',label='上海')
axes[1].plot(x,y2,color='b',linestyle='-.',label='北京')
#显示图例
axes[0].legend()
axes[1].legend()
#x,y刻度
x_label = ['11点{}分'.format(i) for i in x]
axes[0].set_xticks(x[::5])
axes[0].set_xticklabels(x_label[::5])
axes[0].set_yticks(range(0,30,3))
axes[1].set_xticks(x[::5])
axes[1].set_xticklabels(x_label[::5])
axes[1].set_yticks(range(0,30,3))
#网格显示
axes[0].grid(linestyle='--',alpha=0.5)
axes[1].grid(linestyle='--',alpha=0.5)
#图片描述信息
axes[0].set_xlabel('时间')
axes[0].set_ylabel('温度')
axes[0].set_title('上海温度变化')
axes[1].set_xlabel('时间')
axes[1].set_ylabel('温度')
axes[1].set_title('北京温度变化')
#显示图像
plt.show()
```

绘制数学函数图像

```python
#准备x,y
import numpy as np
x= np.linspace(-np.pi,np.pi,10000)#从-pi到pi之间均匀的取10000个数
y=np.sin(x)
#创建画布
plt.figure(figsize=(20,8),dpi=80)
#绘制图像
plt.plot(x,y)
#网格显示
plt.grid(linestyle='--',alpha=0.5)
#显示图像
plt.show()
```

突发奇想绘制椭圆

```python
#准备x,y(采用参数方程的形式绘制椭圆)
import numpy as np
a=np.arange(-2*np.pi,2*np.pi,0.0001)#从-2pi到2pi之间均匀的取数，每个数间隔是0.0001
x=3*np.sin(a)
y=2*np.cos(a)
#创建画布
plt.figure(figsize=(20,8),dpi=90)
#绘制图像
plt.plot(x,y)
#显示网格
plt.grid(linestyle='--',alpha=0.5)
#显示图像
plt.show()
```

同一类不同特征的对比

```python
#准备数据
x=['雷神3：诸神黄昏','正义联盟','寻梦环游记']
y1=[10587.6,10062.5,1275.7]
y2=[36224.9,34479.6,11830]
#创建画布
plt.figure(figsize=(20,8),dpi=100)
#绘制图像
plt.bar(range(len(x)),y1,width=0.2,label='首日票房')
plt.bar([i+0.2 for i in range(len(x))],y2,width=0.2,label='首周票房')
#完善显示
plt.grid(linestyle='--',alpha=0.5)
plt.xticks([i+0.1 for i in range(len(x))],x)
plt.xlabel('电影名')
plt.ylabel('票房（万元/RMB）')
plt.title('电影首日与首周票房对比')
#显示图像
plt.legend()
plt.show()        
```

直方图绘制

```python
#准备数据(电影观影时长)
import random
x=[random.uniform(95,240) for i in range(250)]
#创建画布
plt.figure(figsize=(20,8),dpi=80)
#绘制图像
distance=2#组距
group=int((max(x)-min(x))/2)#组数=（最大值-最小值）/组距
plt.hist(x,bins=group,density=1)
#完善图像
plt.xticks(range(int(min(x)),int(max(x)),distance))
plt.grid(linestyle='--',alpha=0.5)
#显示图像
plt.show()
```
